```python
import numpy as np
import geopandas as gpd
import osm2geojson
import requests
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon
from matplotlib.colors import LinearSegmentedColormap



def IDW(X,Y):
    #Степень "затухания"
    pwr=4
    #вектор-коэффициент обратнопропорциональный расстояниям между расчетной точкой и точками с данными
    distance=1/np.apply_along_axis(lambda x:((X-x[1])**2+(Y-x[0])**2)**0.5,1,data)**pwr
    return((data[:,2]*distance.T).sum(axis=2).T)/distance.sum(axis=0)

#читаем данные (lon-lat-val)
data=np.asarray([[50,37,100],[50,35,200],[50,39,200]])

#делаем запрос к overpass API и возвращаем геометрию 1го полигона!!!
osm_request='''
[out:json];
area[name="Белгородская область"];
rel["admin_level"="4"](area);
out geom;
'''
reg=gpd.GeoDataFrame.from_features(
    osm2geojson.json2geojson(
        requests.get(f"https://maps.mail.ru/osm/tools/overpass/api/interpreter?data={osm_request}").json()
    )['features']
).geometry

#Делаем сетку 1000х1000
x,y=np.meshgrid(
    np.linspace(reg.bounds.minx[0],reg.bounds.maxx[0], 100),
    np.linspace(reg.bounds.miny[0],reg.bounds.maxy[0], 100)
)

#Получаем значения heat для каждой точки сетки
Z=IDW(x,y)

#Рисуем прямоугольник по bounds и вырезаем из него патч по контурам геометрии
fig, ax = plt.subplots(figsize=(25,10))
patch=Polygon(reg[0].geoms[0].exterior.coords,transform=ax.transData)
cmap = LinearSegmentedColormap.from_list(
    'mycmap',
    [
        (0.0, '#660000'),
        (0.15, '#960018'),
        (0.25, '#ff9900'),
        (0.35, '#9acd32'),
        (0.45, '#00a550'),
        (0.55, '#48a1a1'),
        (1.0, '#8b00ff')
    ]
)

im=plt.imshow(
    Z,
    cmap=cmap,
    alpha=0.5,
    extent=(
        reg.bounds.minx[0],
        reg.bounds.maxx[0],
        reg.bounds.miny[0],
        reg.bounds.maxy[0]
    ),
    origin='lower',
    interpolation='bicubic',
    clip_path=patch, clip_on=True
)
plt.colorbar()
plt.show()
```
